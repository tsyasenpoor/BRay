```python
def load_experiment_data(instance_path):
    dfs = {
        "pathways": None,
        "statistics": None,
        "train_stats": None,
        "test_stats": None, 
        "val_stats": None
    }

    if not os.path.isdir(instance_path):
        print(f"Error: Directory not found at {instance_path}")
        return tuple(dfs.values())

    for filename in os.listdir(instance_path):
        file_path = os.path.join(instance_path, filename)
        
        # Skip if not a file
        if not os.path.isfile(file_path):
            continue

        try:
            if 'analysis' in filename and filename.endswith('.json.gz'):
                print(f"Attempting to load {filename} (analysis JSON)...")
                with gzip.open(file_path, 'rt', encoding='utf-8') as f:
                    data = json.load(f)
                dfs['pathways'] = pd.DataFrame(data) 
                print(f"Successfully created DataFrame for {filename}")
            elif 'results' in filename and filename.endswith('.json.gz'):
                with gzip.open(file_path, 'rt', encoding='utf-8') as f:
                    data = json.load(f)
                if isinstance(data, dict):
                    dfs['statistics'] = pd.DataFrame([data])
                elif isinstance(data, list): 
                    dfs['statistics'] = pd.DataFrame(data)
                else:
                    print(f"Warning: statistics file {filename} has an unexpected main data type: {type(data)}. Could not convert to DataFrame.")
            elif 'train' in filename and filename.endswith('.csv.gz'):
                dfs['train_stats'] = pd.read_csv(file_path, compression='gzip')
            elif 'test' in filename and filename.endswith('.csv.gz'):
                dfs['test_stats'] = pd.read_csv(file_path, compression='gzip')
            elif 'val' in filename and filename.endswith('.csv.gz'):
                dfs['val_stats'] = pd.read_csv(file_path, compression='gzip')
        except Exception as e:
            print(f"Error processing file {filename}: {e}")
    return dfs['pathways'], dfs['statistics'], dfs['train_stats'], dfs['test_stats'], dfs['val_stats']
```

```python
def analyze_statistics(statistics_df):
    row_data = statistics_df.iloc[0]
    train_metrics=row_data['train_metrics']
    test_metrics=row_data['test_metrics']
    val_metrics=row_data['val_metrics']

    metrics_list = [
        {'set': 'train', 'metrics': train_metrics},
        {'set': 'validation', 'metrics': val_metrics},
        {'set': 'test', 'metrics': test_metrics}
    ]
    
    metrics_df = pd.DataFrame([
        {'Set': 'Train', **train_metrics},
        {'Set': 'Validation', **val_metrics},
        {'Set': 'Test', **test_metrics}
    ])

    print("Metrics Summary Table:")
    print("=" * 80)
    print(metrics_df.to_string(index=False, float_format='%.4f'))
    print("=" * 80)

    print(f"\nAvailable metrics: {list(train_metrics.keys())}")
    
    return metrics_list, metrics_df
```

```python
import os
import mygene
from matplotlib.patches import Patch
import pandas as pd
import matplotlib.pyplot as plt

# Create the gene_specifics directory inside the experiment folder
gene_specifics_dir = os.path.join(best_exp, "gene_specifics")
os.makedirs(gene_specifics_dir, exist_ok=True)

# Initialize mygene query
mg = mygene.MyGeneInfo()

def get_gene_symbols_batch(gene_ids):
    """Convert Ensembl IDs to gene symbols using mygene"""
    if not gene_ids:
        return {}
    
    # Query mygene for gene symbols
    results = mg.querymany(gene_ids, scopes='ensembl.gene', fields='symbol', species='mouse')
    
    # Create a mapping dictionary
    id_to_symbol = {}
    for result in results:
        if 'symbol' in result:
            id_to_symbol[result['query']] = result['symbol']
        else:
            id_to_symbol[result['query']] = result['query']  # Keep original if no symbol found
    
    return id_to_symbol

# Get top 100 pathways by absolute upsilon value
top_100_pathways = pathway_upsilon_df_abs_sorted.head(100)

print(f"Creating gene activity bar plots for top 100 pathways...")
print(f"Saving plots to: {gene_specifics_dir}")

# Process each pathway
for idx, (pathway_idx, pathway_row) in enumerate(top_100_pathways.iterrows(), 1):
    pathway_name = pathway_row['pathway_name']
    upsilon_val = pathway_row['upsilon']
    
    print(f"Processing {idx}/100: {pathway_name}")
    
    # Get pathway information from best_pathways dataframe
    try:
        pathway_info = best_pathways.loc[pathway_name, 'pathways']
        genes_data = pathway_info['genes']
        
        # Extract gene activities and IDs
        gene_activities = []
        gene_ids = []
        
        for gene in genes_data:
            gene_id = gene.get('gene', 'Unknown')
            if 'activity' in gene:
                activity = gene['activity']
            elif 'weight' in gene:
                activity = gene['weight']
            elif 'coefficient' in gene:
                activity = gene['coefficient']
            else:
                continue
                
            gene_activities.append(activity)
            gene_ids.append(gene_id)
        
        if not gene_activities:
            print(f"  No gene activities found for {pathway_name}")
            continue
        
        # Convert gene IDs to symbols
        id_to_symbol = get_gene_symbols_batch(gene_ids)
        gene_symbols = [id_to_symbol.get(gene_id, gene_id) for gene_id in gene_ids]
        
        # Create a dataframe for easier handling
        gene_df = pd.DataFrame({
            'gene_id': gene_ids,
            'gene_symbol': gene_symbols,
            'activity': gene_activities
        })
        
        # Sort by absolute activity value for better visualization
        gene_df['abs_activity'] = gene_df['activity'].abs()
        gene_df_sorted = gene_df.sort_values('abs_activity', ascending=True)  # ascending for horizontal bars
        
        # Create the plot
        fig, ax = plt.subplots(figsize=(12, max(8, len(gene_df_sorted) * 0.3)))
        
        # Color genes based on positive/negative activity
        colors = ['red' if act < 0 else 'green' for act in gene_df_sorted['activity']]
        
        # Create horizontal bar plot
        bars = ax.barh(range(len(gene_df_sorted)), gene_df_sorted['activity'], 
                       color=colors, alpha=0.7, edgecolor='black', linewidth=0.5)
        
        # Customize the plot
        ax.set_yticks(range(len(gene_df_sorted)))
        ax.set_yticklabels(gene_df_sorted['gene_symbol'], fontsize=8)
        ax.set_xlabel('Gene Activity Level', fontsize=12)
        ax.set_title(f'{pathway_name}\nUpsilon: {upsilon_val:.4f} | Genes: {len(gene_df_sorted)}', 
                    fontsize=10, pad=20)
        
        # Add vertical line at zero
        ax.axvline(x=0, color='black', linestyle='-', alpha=0.8, linewidth=1)
        
        # Add value labels on bars
        for i, (bar, value) in enumerate(zip(bars, gene_df_sorted['activity'])):
            ax.text(value + (0.01 * max(gene_df_sorted['abs_activity']) if value >= 0 else -0.01 * max(gene_df_sorted['abs_activity'])), 
                   bar.get_y() + bar.get_height()/2, 
                   f'{value:.3f}', 
                   ha='left' if value >= 0 else 'right', 
                   va='center', fontsize=6)
        
        # Add legend
        legend_elements = [Patch(facecolor='green', alpha=0.7, label='Positive Activity'),
                          Patch(facecolor='red', alpha=0.7, label='Negative Activity')]
        ax.legend(handles=legend_elements, loc='best', fontsize=10)
        
        # Adjust layout and save
        plt.tight_layout()
        
        # Create safe filename
        safe_filename = pathway_name.replace('/', '_').replace('\\', '_').replace(':', '_')
        safe_filename = safe_filename.replace(' ', '_').replace('(', '').replace(')', '')
        save_path = os.path.join(gene_specifics_dir, f'{idx:03d}_{safe_filename}.png')
        
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        plt.close()
        
        print(f"  Saved: {save_path}")
        
    except Exception as e:
        print(f"  Error processing {pathway_name}: {e}")
        continue

print(f"\nCompleted! All plots saved to {gene_specifics_dir}")

# Create a summary file with pathway information
summary_path = os.path.join(gene_specifics_dir, "pathway_summary.txt")
with open(summary_path, 'w') as f:
    f.write("TOP 100 PATHWAYS BY ABSOLUTE UPSILON VALUE\n")
    f.write("=" * 80 + "\n\n")
    
    for idx, (pathway_idx, pathway_row) in enumerate(top_100_pathways.iterrows(), 1):
        pathway_name = pathway_row['pathway_name']
        upsilon_val = pathway_row['upsilon']
        
        try:
            pathway_info = best_pathways.loc[pathway_name, 'pathways']
            gene_count = len(pathway_info['genes'])
            
            # Create safe filename outside f-string
            safe_name = pathway_name.replace('/', '_').replace('\\', '_').replace(':', '_').replace(' ', '_').replace('(', '').replace(')', '')
            
            f.write(f"{idx:3d}. {pathway_name}\n")
            f.write(f"     Upsilon: {upsilon_val:8.4f}\n")
            f.write(f"     Genes:   {gene_count:3d}\n")
            f.write(f"     File:    {idx:03d}_{safe_name}.png\n\n")
        except:
            f.write(f"{idx:3d}. {pathway_name} - ERROR\n\n")

print(f"Summary file created: {summary_path}")
```
